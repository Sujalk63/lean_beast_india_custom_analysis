```python
df.columns = df.columns.str.strip()
```

```python
df["formatted_posted_time_details"] = pd.to_datetime(
    df["posted_time_details"], utc=True, errors="coerce"
)
df["formatted_posted_time_details"] = df["formatted_posted_time_details"].dt.tz_convert(
    "Asia/Kolkata"
)

df.drop(columns=["posted_time_details"], inplace=True)
```

```python
df["posted_date"] = df["formatted_posted_time_details"].dt.date
```

```python
df.head(5)
```

```python
df.dropna(subset=["likes_count", "comments_count"], inplace=True)
```

```python
df.drop_duplicates(subset=['content_id'], inplace=True)
```

```python
leanbeast_followers = 483000

df["engagement_like_comment"] = df["likes_count"] + df["comments_count"]

df["engagement_per_1000_followers"] = (
    df["engagement_like_comment"] / leanbeast_followers
) * 1000

df["engagement_per_1000_followers"] = df["engagement_per_1000_followers"].round(2)
```

```python
import numpy as np


def parse_duration_to_seconds(duration):
    if pd.isnull(duration):
        return np.nan
    try:
        parts = str(duration).split(":")
        parts = [int(p) for p in parts]
        if len(parts) == 3:  # HH:MM:SS
            return parts[0] * 3600 + parts[1] * 60 + parts[2]
        elif len(parts) == 2:  # MM:SS
            return parts[0] * 60 + parts[1]
        elif len(parts) == 1:  # SS
            return parts[0]
        else:
            return np.nan
    except:
        return np.nan


# Apply conversion
df["video_duration_seconds"] = df["video_duration"].apply(parse_duration_to_seconds)

df[["content_id", "video_duration_seconds"]].head(10)
```

```python
# # Create time period bucket
def time_period(hour):
    if 5 <= hour <= 11:
        return "Morning"
    elif 12 <= hour <= 16:
        return "Afternoon"
    elif 17 <= hour <= 21:
        return "Evening"
    else:
        return "Night"


df["time_period"] = df["hour_of_day"].apply(time_period)


# # Caption length bucket
def caption_length_bucket(length):
    if length <= 32:
        return "Very Short"
    elif length <= 75:
        return "Short"
    elif length <= 326:
        return "Medium"
    else:
        return "Long"


df["caption_length_bucket"] = df["caption_length"].apply(caption_length_bucket)

# Video duration buckets (optional)
df["video_duration_bucket"] = df["video_duration_seconds"].apply(
    lambda x: (
        (
            "Short"
            if x <= 15
            else "Medium" if x <= 30 else "Long" if x <= 60 else "Very Long"
        )
        if pd.notnull(x)
        else x
    )
)
# Export final cleaned and enriched data
# df.to_csv("leanbeast_analysis_ready.csv", index=False)
```

```python
import ast

df['hashtags_used'] = df['hashtags_used'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])
df_exploded = df.explode('hashtags_used')

# Save back to CSV
df_exploded.to_csv("hashtags_exploded.csv", index=False)
```

```python
hashtag_effectivness = (
    df_exploded.groupby(["hashtags_used"])
    .agg(
        median_engagement=("engagement_per_1000_followers", "median"),
        hashtag_count=("content_id", "nunique"),
    )
    .reset_index()
)

hashtag_effectivness.sort_values(by="hashtag_count", ascending=False, inplace=True)
hashtag_effectivness
```
