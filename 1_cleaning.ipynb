```python
import pandas as pd 
```

```python
df = pd.read_excel(r"C:\Users\Sujal Karmakar\Desktop\Desktop\Data Analyst\Python\python_data_analytics_project\lean_beast_india_custom_analysis\scraped_data_lean_beast_india.xlsx")

df.sample(5)
```

```python
df.columns = df.columns.str.strip()
```

```python
df["posted_time_details"].head(2)
```

```python
df["formatted_posted_time_details"] = pd.to_datetime(df["posted_time_details"], utc=True, errors="coerce")
df["formatted_posted_time_details"] = df["formatted_posted_time_details"].dt.tz_convert("Asia/Kolkata")

df.drop(columns=["posted_time_details"], inplace=True)
```

```python
df["posted_date"] = df["formatted_posted_time_details"].dt.date
```

```python
df.head(5)
```

```python
df.dropna(subset=["likes_count", "comments_count"], inplace=True)
```

```python
df.drop_duplicates(subset=['content_id'], inplace=True)
```

```python
leanbeast_followers = 483000

df["engagement_like_comment"] = (df["likes_count"] + df["comments_count"]) 

df["engagement_per_1000_followers"] = (df["engagement_like_comment"] / leanbeast_followers) * 1000

df["engagement_per_1000_followers"] = df["engagement_per_1000_followers"].round(2)

```

```python
df.to_csv("leanbeast_analysis_ready.csv", index=False)
```
